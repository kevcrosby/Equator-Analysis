```python
# Dependencies

import matplotlib.pyplot as plt
import requests as req
import pandas as pd
from citipy import citipy
import numpy as np
import time
import json
import seaborn
import csv
```

# Generate Cities List

```python
#random sampling link
#https://docs.scipy.org/doc/numpy-1.13.0/reference/routines.random.html


sample = 1200
LATs = np.random.uniform(-90,90, size = sample)
LONGs = np.random.uniform(-180,180, size = sample)

COOR = pd.DataFrame({"Latitude":LATs, "Longitude":LONGs})
#COOR.head()

#for rows in COOR:

for index, row in COOR.iterrows():
    city=citipy.nearest_city(row["Latitude"], row["Longitude"])
    COOR.set_value(index, "City", city.city_name)
    COOR.set_value(index, "Country", city.country_code)
    
#COOR.head()

#1200 usualy returns over 500 unique entries
#print(len(COOR["City"].unique()))

#drop repeat cities
COOR = COOR.drop_duplicates(["City"], keep="first")

#final length test:
print("Generated a List of " + str(len(COOR)) + " unique cities.")


#making new columns
COOR["Temperature"] = ""
COOR["Humidity"] = ""
COOR["Cloudiness"] = ""
COOR["Wind Speed"] = ""
```

## Setting Up the API

```python


#API SETUP
apikey = "2e71db067a0204eb9af76572def5ebc3"
url = "http://api.openweathermap.org/data/2.5/weather?q="

```

# Perform API Calls

```python

#setting counters

SetCount = 1
SleepCount = 1


#buiding DF from API CALLS

for index, row in COOR.iterrows():
    
    try:
        #remember to get rid of spaces in city name, 
        #get results in imperial units
        query=url+row["City"].replace(" ","+") + "," + row["Country"] +"&appid="+apikey+"&units=imperial"
        get = req.get(query).json()
        COOR.set_value(index, "Temperature", get["main"]["temp"])
        COOR.set_value(index, "Humidity", get["main"]["humidity"])
        COOR.set_value(index, "Cloudiness", get["clouds"]["all"])
        COOR.set_value(index, "Wind Speed", get["wind"]["speed"])
    except:
        print("Could not find record for that city.")
        COOR.set_value(index, "City", "NOT FOUND")
    
    #increase sleep counter
    SleepCount += 1
    
    
    #SLEEP CONDITIONAL AT 50 SLEEPCOUNT, starts new set
    if SleepCount >= 50:
        print("---Giving API a break to avoid getting blocked---")
        time.sleep(10)
        SetCount += 1
        SleepCount = 1
        
    print("Processing Record " + str(SleepCount) + " of Set " + str(SetCount) + " | " + row["City"])
    print(query)

```

```python
#build dataframes

    #drop all the entries from the dataframe that could not be found in the API
    #marked with city==NOT FOUND
    
OpenWeather_df = COOR.loc[COOR["City"] != "NOT FOUND"]
#OpenWeather_df.head()

#include:
#lat- temp
tempDF = OpenWeather_df[["City","Latitude", "Temperature"]]
#lat - humid
humDF = OpenWeather_df[["City","Latitude", "Humidity"]]
#lat - cloudiness
cloudDF = OpenWeather_df[["City","Latitude", "Cloudiness"]]
#lat - wind speed
windDF = OpenWeather_df[["City","Latitude", "Wind Speed"]]


windDF.head()
```

# Latitude vs Temperature Plot

```python
#lat vs Temp Scatter Plot

plt.scatter(tempDF["Latitude"], tempDF["Temperature"], color="c0")
plt.title("Temperature vs Latitude")
plt.xlabel("Latitude")
plt.ylabel("Temperature in Farenheit")
plt.grid( linestyle='-')
plt.show()
```
